# countervector & hashing vector  for topic modeling

# 1.countervector

```python
import pandas as pd
```

```python
df = pd.read_csv("WhatsApp Chat with Ravi Sir.txt",header=None,error_bad_lines=False,encoding='utf8')
```

```python
df
```

```python
# to delet the column 2,3

# Display the column names
print("Column names:")
print(df.columns)

# Display the original DataFrame (first few rows for inspection)
print("Original DataFrame:")
print(df.head())

# Assuming the actual column names are '2' and '3' (as strings)
# If they are integers, you should use integers instead
columns_to_delete = [2, 3]

# Check if columns exist and delete them
columns_to_delete = [col for col in columns_to_delete if col in df.columns]
df.drop(columns=columns_to_delete, inplace=True)
print(df.head())
```

```python
df.columns=['Date','Chat']
Message=df["Chat"].str.split("-",n=1,expand=True)
df["Time"]=Message[0]
Message1=Message[1].str.split(":",n=1,expand=True)
df["Name"]=Message1[0]
df["Chat"]=Message1[1]
df=df[["Date","Time","Name","Chat"]]
df
```

```python
#to delete the first row for unknown data
df = df.iloc[1:]

# Optionally, reset the index if needed
df.reset_index(drop=True, inplace=True)
df
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize the text data
count_vectorizer = CountVectorizer(stop_words='english')
count_data = count_vectorizer.fit_transform(df['Chat'])

# Fit the LDA model
num_topics = 5
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(count_data)

# Display the top words for each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Number of top words to display
no_top_words = 5

# Use the appropriate method based on the version of scikit-learn
try:
    feature_names = count_vectorizer.get_feature_names_out()
except AttributeError:
    feature_names = count_vectorizer.get_feature_names()

display_topics(lda, feature_names, no_top_words)

# Assign topics to documents
lda_transformed = lda.transform(count_data)
df['topic'] = lda_transformed.argmax(axis=1)

print(df)
```

# hashing vector topic modeling

```python
dataset=df
```

```python
dataset
```

```python
import pandas as pd
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize the text data using HashingVectorizer
hash_vectorizer = HashingVectorizer(stop_words='english', n_features=2**10, alternate_sign=False)
hash_data = hash_vectorizer.fit_transform(dataset['Chat'])

# Fit the LDA model
num_topics = 4
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(hash_data)

# Since HashingVectorizer does not provide feature names, we need to simulate them
# Here, we will use placeholder names for simplicity
feature_names = [f'feature_{i}' for i in range(hash_data.shape[1])]

# Display the top words for each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Number of top words to display
no_top_words = 5
display_topics(lda, feature_names, no_top_words)

# Assign topics to documents
lda_transformed = lda.transform(hash_data)
dataset['topic1'] = lda_transformed.argmax(axis=1)

print(dataset)

```
