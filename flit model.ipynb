```python
import pandas as pd
```

```python
df = pd.read_csv("flirting_rated.csv",header=None,error_bad_lines=False,encoding='utf8')
```

```python
df= pd.DataFrame(df)
df
```

```python

# Rename the columns
df.rename(columns={0: 'label', 1: 'chat'}, inplace=True)

# Print the DataFrame with renamed columns
df = df[['chat', 'label']]
#to delete the first row for unknown data
df = df.iloc[1:]

# Optionally, reset the index if needed
df.reset_index(drop=True, inplace=True)
df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Splitting dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df['chat'], df['label'], test_size=0.2, random_state=42)

# Creating Counter Vectorizer
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Initializing and training Logistic Regression model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Making predictions
predictions = model.predict(X_test_vec)

# Evaluating the model
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```
